```python
X = df.iloc[:,df.columns != "Exited"]
```

```python
y = df["Exited"]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test ,y_train , y_test = train_test_split(X,y,test_size = 0.2 , random_state = 2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
```

```python
X_train=Scaler.fit_transform(X_train)
X_test =Scaler.transform(X_test)
```

### logistic REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import  accuracy_score,confusion_matrix , f1_score , precision_score , recall_score
```

```python
acc= accuracy_score(y_test , y_pred)
f1 = f1_score(y_test , y_pred)
prec = precision_score(y_test , y_pred)
rec = recall_score(y_test , y_pred)
```

```python
result = pd.DataFrame([['Logistic Regression ',acc,f1,prec,rec]],
                      columns = ['Model','Accuracy','F1 Score','Precision','Recall'])
```

```python
result
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy = cross_val_score(estimator = lr,X=X_train,y=y_train,cv=10)
```

```python
Accuracy1= format(accuracy.mean()*100)
print(f"Accuracy score of Logistic Regression  is : {Accuracy1}")
print("Standard Deviation is : {:.2f} %".format(accuracy.std()*100))
```
